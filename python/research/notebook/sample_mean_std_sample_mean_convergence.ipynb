```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define sample mean and standardized sample mean calculation for normal and binomial distributions

```python
def sample_mean_normal(size, mu, sigma, n1):
    # size - number of samples to draw for sample mean calculation
    # mu - mean of the population distribution
    # sigma - std of the population distribution
    # n1 - number of sample means to calculate, a total of size * n1 samples will be drawn
    
    res = []
    for i in range(n1):
        res.append(np.mean(np.random.normal(mu, sigma, size)))
    return res


def sample_mean_binomial(size, n, p, n1):
    # size - number of samples to draw for sample mean calculation
    # mu - mean of the population distribution
    # sigma - std of the population distribution
    # n1 - number of sample means to calculate, a total of n * n1 samples will be drawn
    
    res = []
    for i in range(n1):
        res.append(np.mean(np.random.binomial(n, p, size)))
    return res

def standard_sample_mean_normal(size, mu, sigma, n1):
    # size - number of samples to draw for sample mean calculation
    # mu - mean of the population distribution
    # sigma - std of the population distribution
    # n1 - number of standard sample means to calculate, a total of size * n1 samples will be drawn
    
    res = []
    for i in range(n1):
        res.append(np.sqrt(size) * (np.mean(np.random.normal(mu, sigma, size)) - mu)/sigma)
    return res


def standard_sample_mean_binomial(size, n, p, n1):
    # size - number of samples to draw for sample mean calculation
    # mu - mean of the population distribution
    # sigma - std of the population distribution
    # n1 - number of standard sample means to calculate, a total of n * n1 samples will be drawn
    
    mu = n * p
    sigma = np.sqrt(n * p * (1-p))
    
    res = []
    for i in range(n1):
        res.append(np.sqrt(size) * (np.mean(np.random.binomial(n, p, size)) - mu)/sigma)
    return res
```

### Plot the sampling distributions of the sample mean for n samples drawn from the normal distribution

```python
fig, axs = plt.subplots(1, 4)

fig.set_figwidth(18)
fig.set_figheight(6)
mu = 0
sigma = 0.1
n1 = 1000
bins = 20

for idx, i in enumerate([1, 10, 100, 1000]):
    ri = sample_mean_normal(i, mu, sigma, n1)
    axs[idx].hist(ri, orientation="horizontal", bins=20)
    axs[idx].set_ylim(mu - 4*sigma, mu + 4*sigma)
    
    axs[idx].spines['top'].set_visible(False)
    axs[idx].spines['right'].set_visible(False)
    axs[idx].spines['bottom'].set_visible(False)

plt.show()
```

### Plot the sampling distributions of the standardized sample mean for n samples drawn from the normal distribution

```python
fig, axs = plt.subplots(1, 4)

fig.set_figwidth(18)
fig.set_figheight(6)

mu = 0
sigma = 0.1
n1 = 1000
bins = 20

for idx, i in enumerate([1, 10, 100, 1000]):
    ri = standard_sample_mean_normal(i, mu, sigma, n1)
    axs[idx].hist(ri, orientation="horizontal", bins=20)
    axs[idx].set_ylim(-4, 4)
    
    axs[idx].spines['top'].set_visible(False)
    axs[idx].spines['right'].set_visible(False)
    axs[idx].spines['bottom'].set_visible(False)

plt.show()
```
